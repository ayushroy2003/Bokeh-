Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Ans = In bokeh glyphs are the basic visual element used to create plot . they represent shapes like circle square, and line to visulize data points


```python
from bokeh.plotting import figure , output_file ,show
from bokeh.io import output_notebook

output_notebook()

x = [1,2,3,4,5]
y1= [6,7,8,9,10]
y2 = [2,3,5,7,6]
y3 = [5,6,2,3,4]

p = figure(title= "Glyphs in Bokeh",x_axis_label='x' ,y_axis_label='y')

p.circle(x,y1, size=10 , color="navy",alpha=0.5 ,legend_label="CIRCLE")

p.square(x,y2, size=10, color="green" ,alpha=0.5 ,legend_label="SQUARE")

p.triangle(x,y3 ,size= 10,color="red" ,alpha= 0.5 ,legend_label="TRIANGLE")

show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

Ans = 

```python
from bokeh.plotting import figure , output_file , show
from bokeh.io import output_notebook

output_notebook()

x = [1,2,3,4,5]
y = [6,7,8,9,10]

p = figure(title= "glyphs in bokeh" )
p.xaxis.axis_label = "X AXIS"
p.yaxis.axis_label = "Y AXIS"

p.line(x=x ,y=y ,line_width=3 , line_color = 'red' , legend_label="DATA POINTS")
p.circle(x=x, y=y , size=15, fill_color='green')


show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

Ans - Bokeh server allows you to build interactive web applications that can update plots and data in real time . it provides the capability to stream data to a plot , respond to user interactions and connect to live data source 

Key Features of a Bokeh Server

- Real time Update : THe bokeh server enables plots and data to be updated dynamically based on new incoming data . this is particularly usefl for application like live data monitering or interactive dashboards

- Interactive Widgets : Use sliders , dropdowns and other widget to control plots 

- Callback Functions : Define custom behaviour for user interactions 

- Data Synchronization : Keeps client and server data in sync

How to use a Bokeh Server 

1. Define the application : Set up data source, plots and widgets into a coherent layout 

2. Create a Layout : Organize plots and widgets into a cohernt layout

3. Add Callbacks : implement function to update plots based on interaction or new data

4. Run the server : use 'bokeh serve' to host the application 



Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

ANs - 

to embed a Bokeh plot into a web page or dashboard using Flask or Django, you can use the Bokeh server together with either of these web frameworks. Here are the basic steps:

Define a Bokeh application that creates the desired plot and any other interactive elements you want to include.
Start the Bokeh server from your Flask or Django application, using the bokeh.server.server.Server class.
Create a URL route in your Flask or Django application that serves as the entry point for the Bokeh app.
In the Bokeh app, use the bokeh.embed.server_document function to generate an HTML script tag that loads the Bokeh app into the web page.

```python
from flask import Flask, render_template
from bokeh.embed import components
from bokeh.plotting import figure
from bokeh.resources import CDN

app = Flask(__name__)

@app.route('/')
def index():
    # Create a Bokeh plot
    plot = figure(title="Simple Bokeh Plot", x_axis_label='x', y_axis_label='y')
    plot.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], legend_label="Temp.", line_width=2)

    # Embed plot into HTML via Bokeh components
    script, div = components(plot)
    cdn_js = CDN.js_files[0]
    cdn_css = CDN.css_files[0]

    return render_template('index.html', script=script, div=div, cdn_js=cdn_js, cdn_css=cdn_css)

if __name__ == '__main__':
    app.run(debug=True)
```
